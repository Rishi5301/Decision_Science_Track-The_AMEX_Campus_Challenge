```python
# Calculate subplot dimensions
n_columns = len(interest_column)
n_rows = int(np.ceil(n_columns / 3))  # 3 columns per row, adjust as needed
n_cols = min(3, n_columns)  # Max 3 columns

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))

# Handle case where there's only one subplot
if n_columns == 1:
    axes = [axes]
elif n_rows == 1:
    axes = axes.reshape(1, -1)

# Flatten axes array for easier indexing
axes_flat = axes.flatten()

# Plot each histogram
for i, column in enumerate(interest_column):
    clean_data = df[column].dropna()
    
    # Plot histogram
    axes_flat[i].hist(clean_data, bins=17, edgecolor='black')
    axes_flat[i].set_title(f'Graph of {column}')
    axes_flat[i].set_xlabel('Index')
    axes_flat[i].set_ylabel(column)
    axes_flat[i].tick_params(axis='x', rotation=90)
    
    # Print statistics
    print(f'Column: {column}')
    print(f'  Number of NONE values: {df[column].isna().sum()}')
    print(f'  In percentage: {df[column].isna().sum()/df.shape[0]:.2%}')
    print(f'  Median: {df[column].median()}')
    print('-' * 40)

# Hide extra subplots if any
for i in range(n_columns, len(axes_flat)):
    axes_flat[i].set_visible(False)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
